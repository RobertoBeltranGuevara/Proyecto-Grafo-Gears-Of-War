```python
# Obtener el grado de todos los nodos
grados = dict(G.degree())
for nodo, grado in grados.items():
    print(f"El grado del nodo '{nodo}' es: {grado}")
```

```python
# Calcular la suma de los grados de los nodos
suma_grados = sum(dict(G.degree()).values())

# Contar el número de aristas
num_aristas = G.number_of_edges()

# Verificar el Teorema del apretón de manos (Handshaking Lemma)
if suma_grados == 2 * num_aristas:
    print(f"La suma de los grados de los nodos ({suma_grados}) es igual a 2 veces el número de aristas ({2 * num_aristas}).")
else:
    print(f"La suma de los grados de los nodos ({suma_grados}) no es igual a 2 veces el número de aristas ({2 * num_aristas}).")
```

```python
# Calcular la excentricidad de todos los nodos
excentricidades = nx.eccentricity(G)

# Mostrar la excentricidad de cada nodo
for nodo, excentricidad in excentricidades.items():
    print(f"La excentricidad del nodo '{nodo}' es: {excentricidad}")
```

```python
# Calcular radio, diámetro y centro
radio = min(excentricidades.values())
diametro = max(excentricidades.values())
centro = [nodo for nodo, excentricidad in excentricidades.items() if excentricidad == radio]

# Mostrar los resultados
print(f"Radio del grafo: {radio}")
print(f"Diámetro del grafo: {diametro}")
print(f"Centro del grafo: {centro}")
```

```python
pos = nx.circular_layout(G)

# Dibujar el grafo
plt.figure(figsize=(12, 12))
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_size=8, font_color='black', node_size=500, edge_color='gray')
plt.title('Grafo de Gears of War')
plt.show()
```

```python
# Verificar si el grafo tiene un camino euleriano
if nx.is_eulerian(G):
    print("El grafo tiene un camino euleriano.")
    camino_euleriano = list(nx.eulerian_circuit(G))
    print("Camino Euleriano:")
    print(camino_euleriano)
    
    # Dibujar el grafo con el camino euleriano resaltado
    plt.figure(figsize=(12, 12))
    pos = nx.spring_layout(G, seed=42)
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")
    nx.draw_networkx_edges(G, pos, edgelist=camino_euleriano, edge_color="red", width=3)
    plt.title("Grafo de Gears of War con Camino Euleriano")
    plt.show()
else:
    print("El grafo no tiene un camino euleriano.")
```
